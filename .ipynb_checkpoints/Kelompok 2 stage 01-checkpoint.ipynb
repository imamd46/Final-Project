### Korelasi Variabel kategori  dengan Target

```python
from scipy.stats import chi2_contingency
```

```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x,y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))
```

```python
plt.figure(figsize=(15,10))
for x in range(0, len(cats)):
  plt.subplot(3, 3, x+1)
  corr = cramers_v(df[cats[x]], df["TravelInsurance"])
  sns.heatmap([[1, corr], [corr, 1]], annot=True)
  plt.title(cats[x], fontsize = 10)
  plt.tight_layout()
```

```python
# Calculate Cramer's V matrix for all categorical variables
cramer_v_matrix = np.zeros((len(cats), len(cats)))
for i in range(len(cats)):
    for j in range(len(cats)):
        cramer_v_matrix[i,j] = cramers_v(df[cats[i]], df[cats[j]])

# Plot Cramer's V matrix using heatmap
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(cramer_v_matrix, ax=ax, annot=True, xticklabels=cats, yticklabels=cats, cmap="YlGnBu")
```

Variabel kategori Heatmap yang berkorelasi dengan fitur target:
- variabel EverTravelledAbroad, FrequentFlyer dan Employment Type

Variabel kategori Heatmap yang kemungkinan berkorelasi dengan fitur lainnya:
- variabel EverTravelledAbroad dengan FrequentFlyer
- variabel EverTravelledAbroad dengan Employment Type
- variabel Employment Type dengan GraduateOrNot
- variabel FrequentFlyer dengan Employment Type

```python
data_emp_ins = pd.crosstab(df['Employment Type'],df['TravelInsurance'])
data_emp_ins['Total'] = data_emp_ins.sum(axis=1)

sns.set_style("whitegrid")
ax = sns.barplot(x=data_emp_ins.index, y=data_emp_ins[1], color='b')
ax.set_title('Jumlah Responden yang Memiliki Travel Insurance Berdasarkan Jenis Pekerjaan')
ax.set_xlabel('Jenis Pekerjaan')
ax.set_ylabel('Jumlah Responden')

# Legend jumlah responden
for i in range(data_emp_ins.shape[0]):
    ax.text(i, data_emp_ins[1][i] + 5, str(data_emp_ins[1][i]), ha='center', va='bottom')

# Legend persentase jumlah responden
for i in range(data_emp_ins.shape[0]):
    ax.text(i, data_emp_ins[1][i]/2, str(round(data_emp_ins[1][i]/data_emp_ins['Total'][i]*100,2))+'%', ha='center', va='bottom', fontdict={'fontsize': 10})

plt.show()
```

**Jenis pekerjaan sangat mempengaruhi tingkat kepemilikan travel insurance. Responden dengan jenis pekerjaan "Private Sector/Self Employed" merupakan kelompok dengan jumlah responden yang paling banyak memiliki travel insurance, sedangkan responden dengan jenis pekerjaan "Government Sector" merupakan kelompok dengan jumlah responden yang paling sedikit memiliki travel insurance.**
